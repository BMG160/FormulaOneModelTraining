```python
#Merging all data tables into one main dataset
import pandas as pd

drivers = pd.read_csv("drivers.csv")
constructors = pd.read_csv("constructors.csv")
races = pd.read_csv("races.csv")
results = pd.read_csv("results.csv")
status = pd.read_csv("status.csv")

df = results.merge(drivers, on="driverId", how="left")
df = df.merge(constructors, on="constructorId", how="left")
df = df.merge(races, on="raceId", how="left")
df = df.merge(status, on="statusId", how="left")

print(df.info())
print(df.isnull().sum())

df.to_csv("f1_merged_dataset.csv", index=False)
print("Core merged dataset saved successfully.")
```

```python
import pandas as pd
import numpy as np

fe_df = df.copy()

print("Columns in merged dataset:")
print(fe_df.columns.tolist())


if "date" in fe_df.columns:
    fe_df["date"] = pd.to_datetime(fe_df["date"], format="%d/%m/%Y", errors="coerce")

if "dob" in fe_df.columns:
    fe_df["dob"] = pd.to_datetime(fe_df["dob"], format="%Y-%m-%d", errors="coerce")


fe_df = fe_df.sort_values(["year", "round", "raceId", "driverId"]).reset_index(drop=True)


if "race_points" in fe_df.columns:
    points_col = "race_points"
elif "points_x" in fe_df.columns:
    points_col = "points_x"
elif "points" in fe_df.columns:
    points_col = "points"
elif "points_y" in fe_df.columns:
    points_col = "points_y"
else:
    raise KeyError("No points column found. Check your merged dataset columns.")

print("Using points column:", points_col)

rename_map = {}

if "name_x" in fe_df.columns:
    rename_map["name_x"] = "constructor_name"

if "name_y" in fe_df.columns:
    rename_map["name_y"] = "race_name"

if "nationality_x" in fe_df.columns:
    rename_map["nationality_x"] = "driver_nationality"

if "nationality_y" in fe_df.columns:
    rename_map["nationality_y"] = "constructor_nationality"

if "time_x" in fe_df.columns:
    rename_map["time_x"] = "result_time"

if "time_y" in fe_df.columns:
    rename_map["time_y"] = "race_time"

fe_df = fe_df.rename(columns=rename_map)

if points_col != "race_points":
    fe_df = fe_df.rename(columns={points_col: "race_points"})
    points_col = "race_points"

numeric_cols = [
    "grid", "positionOrder", "race_points", "laps", "milliseconds",
    "fastestLap", "rank", "year", "round", "driverId", "constructorId", "circuitId"
]

for col in numeric_cols:
    if col in fe_df.columns:
        fe_df[col] = pd.to_numeric(fe_df[col], errors="coerce")


if "date" in fe_df.columns and "dob" in fe_df.columns:
    fe_df["driver_age"] = ((fe_df["date"] - fe_df["dob"]).dt.days / 365.25)

if "positionOrder" in fe_df.columns:
    fe_df["is_win"] = (fe_df["positionOrder"] == 1).astype(int)
    fe_df["is_podium"] = (fe_df["positionOrder"] <= 3).astype(int)

if "status" in fe_df.columns:
    fe_df["finished_race"] = fe_df["status"].astype(str).str.contains("Finished", case=False, na=False).astype(int)

if "grid" in fe_df.columns and "positionOrder" in fe_df.columns:
    fe_df["positions_gained"] = fe_df["grid"] - fe_df["positionOrder"]

fe_df["driver_experience"] = fe_df.groupby("driverId").cumcount()
fe_df["constructor_experience"] = fe_df.groupby("constructorId").cumcount()
fe_df["driver_circuit_experience"] = fe_df.groupby(["driverId", "circuitId"]).cumcount()

fe_df["driver_avg_finish_before_race"] = (
    fe_df.groupby("driverId")["positionOrder"]
    .transform(lambda x: x.shift(1).expanding().mean())
)

fe_df["driver_avg_points_before_race"] = (
    fe_df.groupby("driverId")["race_points"]
    .transform(lambda x: x.shift(1).expanding().mean())
)

fe_df["driver_points_last_3"] = (
    fe_df.groupby("driverId")["race_points"]
    .transform(lambda x: x.shift(1).rolling(3, min_periods=1).mean())
)

fe_df["driver_points_last_5"] = (
    fe_df.groupby("driverId")["race_points"]
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)

fe_df["driver_finish_last_3"] = (
    fe_df.groupby("driverId")["positionOrder"]
    .transform(lambda x: x.shift(1).rolling(3, min_periods=1).mean())
)

fe_df["driver_finish_last_5"] = (
    fe_df.groupby("driverId")["positionOrder"]
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)

fe_df["previous_race_finish"] = fe_df.groupby("driverId")["positionOrder"].shift(1)
fe_df["previous_race_points"] = fe_df.groupby("driverId")["race_points"].shift(1)
fe_df["previous_grid"] = fe_df.groupby("driverId")["grid"].shift(1)

fe_df["constructor_avg_finish_before_race"] = (
    fe_df.groupby("constructorId")["positionOrder"]
    .transform(lambda x: x.shift(1).expanding().mean())
)

fe_df["constructor_avg_points_before_race"] = (
    fe_df.groupby("constructorId")["race_points"]
    .transform(lambda x: x.shift(1).expanding().mean())
)

fe_df["constructor_points_last_3"] = (
    fe_df.groupby("constructorId")["race_points"]
    .transform(lambda x: x.shift(1).rolling(3, min_periods=1).mean())
)

fe_df["constructor_points_last_5"] = (
    fe_df.groupby("constructorId")["race_points"]
    .transform(lambda x: x.shift(1).rolling(5, min_periods=1).mean())
)

fe_df["season_driver_points_before_race"] = (
    fe_df.groupby(["year", "driverId"])["race_points"]
    .transform(lambda x: x.shift(1).cumsum())
)

fe_df["season_constructor_points_before_race"] = (
    fe_df.groupby(["year", "constructorId"])["race_points"]
    .transform(lambda x: x.shift(1).cumsum())
)

fe_df["season_race_number_for_driver"] = (
    fe_df.groupby(["year", "driverId"]).cumcount() + 1
)

fe_df["season_race_number_for_constructor"] = (
    fe_df.groupby(["year", "constructorId"]).cumcount() + 1
)

fe_df["driver_avg_finish_at_circuit"] = (
    fe_df.groupby(["driverId", "circuitId"])["positionOrder"]
    .transform(lambda x: x.shift(1).expanding().mean())
)

fe_df["driver_avg_points_at_circuit"] = (
    fe_df.groupby(["driverId", "circuitId"])["race_points"]
    .transform(lambda x: x.shift(1).expanding().mean())
)

zero_fill_cols = [
    "driver_experience",
    "constructor_experience",
    "driver_circuit_experience",
    "season_driver_points_before_race",
    "season_constructor_points_before_race",
    "previous_race_points"
]

for col in zero_fill_cols:
    if col in fe_df.columns:
        fe_df[col] = fe_df[col].fillna(0)

median_fill_cols = [
    "driver_avg_finish_before_race",
    "driver_finish_last_3",
    "driver_finish_last_5",
    "previous_race_finish",
    "constructor_avg_finish_before_race",
    "driver_avg_finish_at_circuit"
]

for col in median_fill_cols:
    if col in fe_df.columns:
        fe_df[col] = fe_df[col].fillna(fe_df["positionOrder"].median())

points_fill_cols = [
    "driver_avg_points_before_race",
    "driver_points_last_3",
    "driver_points_last_5",
    "constructor_avg_points_before_race",
    "constructor_points_last_3",
    "constructor_points_last_5",
    "driver_avg_points_at_circuit"
]

for col in points_fill_cols:
    if col in fe_df.columns:
        fe_df[col] = fe_df[col].fillna(0)

if "previous_grid" in fe_df.columns:
    fe_df["previous_grid"] = fe_df["previous_grid"].fillna(fe_df["grid"].median())

if "driver_age" in fe_df.columns:
    fe_df["driver_age"] = fe_df["driver_age"].fillna(fe_df["driver_age"].median())

drop_cols = [
    "resultId",
    "driverRef",
    "constructorRef",
    "url_x",
    "url_y",
    "dob",
    "date",
    "forename",
    "surname",
    "code",
    "result_time",
    "race_time"
]

existing_drop_cols = [col for col in drop_cols if col in fe_df.columns]
fe_df = fe_df.drop(columns=existing_drop_cols)

print("Feature engineered dataset shape:", fe_df.shape)
print("\nTop missing values:")
print(fe_df.isnull().sum().sort_values(ascending=False).head(20))

print("\nSample rows:")
print(fe_df.head())

fe_df.to_csv("f1_feature_engineered_dataset.csv", index=False)
print("\nFeature engineering complete. File saved as f1_feature_engineered_dataset.csv")
```

```python
#Setting target value and label encoding
import pandas as pd
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("f1_feature_engineered_dataset.csv", low_memory=False)

print("Dataset shape:", df.shape)

target = "positionOrder"

y = df[target]
X = df.drop(columns=[target])

print("Target column:", target)

categorical_cols = X.select_dtypes(include=["object"]).columns.tolist()

print("\nCategorical columns:")
print(categorical_cols)

label_encoders = {}

for col in categorical_cols:
    le = LabelEncoder()
    X[col] = X[col].astype(str)
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le

print("\nCategorical encoding completed.")

final_df = pd.concat([X, y], axis=1)

print("\nFinal dataset shape:", final_df.shape)

print("\nRemaining missing values:")
print(final_df.isnull().sum().sort_values(ascending=False).head(20))

final_df.to_csv("f1_ml_ready_dataset.csv", index=False)

print("\nML-ready dataset saved as:")
print("f1_ml_ready_dataset.csv")
```

```python
#Data splitting for testing and training
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("f1_ml_ready_dataset.csv", low_memory=False)

print("Dataset shape:", df.shape)

target = "positionOrder"

X = df.drop(columns=[target])
y = df[target]

print("Feature shape:", X.shape)
print("Target shape:", y.shape)

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,      # 80% train, 20% test
    random_state=42
)

print("\nTrain/Test split completed")

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)

print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)

X_train.to_csv("X_train.csv", index=False)
X_test.to_csv("X_test.csv", index=False)

y_train.to_csv("y_train.csv", index=False)
y_test.to_csv("y_test.csv", index=False)

print("\nDatasets saved:")
print("X_train.csv")
print("X_test.csv")
print("y_train.csv")
print("y_test.csv")
```
